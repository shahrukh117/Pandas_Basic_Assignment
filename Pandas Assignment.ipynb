### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

A DataFrame in pandas is a two-dimensional labeled data structure that can store data of different types (e.g., numbers, strings, etc.) in columns. It is similar to a table or a spreadsheet where each column represents a different variable, and each row represents a different observation.

On the other hand, a pandas Series is a one-dimensional labeled data structure that can store data of the same type. It is like a single column of a DataFrame.


```python
import pandas as pd

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Claire'],
    'Age': [25, 30, 27],
    'Gender': ['Female', 'Male', 'Female']
}
df = pd.DataFrame(data)

# Creating a Series
ages = pd.Series([25, 30, 27])

# Printing the DataFrame
print("DataFrame:")
print(df)

# Printing the Series
print("\nSeries:")
print(ages)
```

### Q7. Create a DataFrame using multiple Series. Explain with an example.

You can create a DataFrame using multiple Series by passing them as a dictionary to the DataFrame constructor. Each Series will become a column in the DataFrame.Here's an example:

```python
import pandas as pd

name_series = pd.Series(['Anuj', 'Rahul', 'Manoj'])
age_series = pd.Series([25, 30, 35])
city_series = pd.Series(['New Delhi', 'Noida', 'Gaziabad'])

data = {'Name': name_series, 'Age': age_series, 'City': city_series}

df = pd.DataFrame(data)
print(df)
```

In this example, we have three Series: `name_series`, `age_series`, and `city_series`. We create a dictionary called `data` where the keys are the column names ('Name', 'Age', 'City') and the values are the respective Series. Finally, we pass the `data` dictionary to the DataFrame constructor to create the DataFrame `df`. Each Series becomes a column in the DataFrame
